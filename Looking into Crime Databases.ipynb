# Section 2: Exploratory Data Analysis
To give me a broad overview of the data, I can do an analysis of the descriptive statistics for certain areas of the data. For example, I can use the "TIME OCC" column to find the average time of occurrence, along with the standard deviation. Using these results, I can further analyze if there is a specific time of day where crime may occur more often. 

Beyond descriptive statistics, I can see which description of the case (type of crime) occurs more often than others. To do this, I will use 'value_counts()' and find the most common type of crime case. To combat this, I will also use 'value_counts()' to find the least common type of case, using idmin() instead of idmax(). Seeing what the least common type of case is, I may be able to change my mind about which forms of crime are most common and which are not.

```python
average = df['TIME OCC'].mean()
average
```

```python
from statistics import median
med = median(df['TIME OCC'])
med
```

```python
standarddeviation = df['TIME OCC'].std()
standarddeviation
```

```python
from statistics import mode
modee = mode(df['TIME OCC'])
modee
```

With these three descriptive statistics--mean, median, standard deviation, and mode--we can analyze the shape of the data. With a mean of around 1336 (1:36 pm) and a median of about 1415 (2:15 pm), we can understand that the data has a left-skew. This means that the majority of the data is above the mean, which shows that this data is not normally distributed. However, we have a standard deviation of 654 (6 hours and 54 minutes), which can show us that our data has high variability. Finally, using mode, we can see that the most common time for crime to occur is 12pm, which is surprising. Expecting nighttime to be more dangerous, this data can show that crime happens at all hours of the day.

```python
most_common_value = df['Crm Cd Desc'].value_counts().idxmax()
most_common_value
```

```python
least_common_value = df['Crm Cd Desc'].value_counts().idxmin()
least_common_value
```

The most common crime case type in this dataset seems to be stolen vehicle. I expected some form of aggravated assault or store robberies to be the most common forms of crime in the system, but this was able to prove me wrong. Inciting a riot stands as the least common form of crime in this dataset. Although riots have happened a lot in the past few years, I still am not surprised this would be the least common form. When there are riots, it is hard to find not only who started it, but who actually joined in and organized it.

# Section 4: Data Science
I am curious to see if there is any correlation between the area the crime occurred and the time it occurred at. As I explained earlier, I expected crime to occur later at night rather than during the day. My analysis earlier proved that theory to be incorrect, but I wonder if that just fluctuates throughout the neighborhoods. In using Python, I can find the correlation between 'AREA' and 'TIME OCC' to answer my questions.

```python
correlation_matrix = df.corr(numeric_only = True)
correlation_coefficient = correlation_matrix.loc['AREA', 'TIME OCC']
correlation_coefficient
```

Our correlation coefficient between Area and Time OCC seems to be around 0. This signifies no correlation between the two variables, which combats my theory that time of crimes fluctuates throughout neighborhoods. There is no relationship between the time a crime occurs and the area it occurs in.
